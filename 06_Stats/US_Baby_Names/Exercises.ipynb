### Step 1. Import the necessary libraries

```python
import pandas as pd 
import numpy as mp 
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv). 

```python
baby_names = pd.read_csv("US_Baby_Names_right.csv")
```

### Step 3. Assign it to a variable called baby_names.

```python
baby_names.head(10)
```

### Step 4. See the first 10 entries

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names = baby_names.drop(columns=['Unnamed: 0','Id'])
```

### Step 6. What year has the highest number of baby names in the dataset?

```python
baby_names_agg = baby_names.groupby(by = 'Year').count()['Name'].reset_index().sort_values(by = 'Name')
baby_names_agg
```

```python
baby_names_agg.loc[baby_names_agg['Name'].idxmax(),'Year']
```

### Step 7. Is there more male or female names in the dataset?

```python
def is_male(x):
    return (x == 'M').sum()

def is_female(x):
    return (x == 'F').sum()


baby_names.groupby(lambda x: 'Total').agg(
    Male = ('Gender',is_male),
    Female = ('Gender',is_female),
)
```

```python
(baby_names.Gender == 'M').sum()
```

### Step 8. Group the dataset by name and assign to names

```python
names = baby_names.groupby(by = 'Name').agg(
    Count = ('Count',pd.Series.sum)
).reset_index().sort_values(by = 'Count',ascending=False)

names
```

### Step 9. How many different names exist in the dataset?

```python
names['Name'].nunique()
```

### Step 10. What is the name with most occurrences?

```python
names.loc[names['Count'].idxmax()]
```

### Step 11. How many different names have the least occurrences?

```python
names.loc[names['Count'] ==5]
```

### Step 12. What is the median name occurrence?

```python
names['Count'].median()
```

### Step 13. What is the standard deviation of names?

```python
names['Count'].std()
```

### Step 14. Get a summary with the mean, min, max, std and quartiles.

```python
names['Count'].describe()
```
