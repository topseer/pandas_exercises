```python
import pandas as pd
import numpy as np
```

```python
# 1. Simple dict — keys become columns
d = {'name': ['Alice', 'Bob'], 'age': [30, 25]}
pd.DataFrame(d)
```

```python
# 2. Dict of scalars — use .items() for key-value pairs
d = {'CA': 'West', 'NY': 'East', 'TX': 'South'}
pd.DataFrame(d.items(), columns=['state', 'region'])
```

```python
# 3. List of dicts — each dict becomes a row
d = [{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 25}]
pd.DataFrame(d)
```

```python
# 4. Nested dict — outer keys become columns, inner keys become index
d = {'Alice': {'math': 90, 'english': 85}, 'Bob': {'math': 70, 'english': 95}}
df = pd.DataFrame(d)
df
```

```python
for i, row in df.iterrows():
    print (row['Alice'])
    print (row.Alice)
    print ('i=',i)
```
