### Simple Random Split (50/50)

```python
import pandas as pd

df = pd.read_csv("US_Crime_Rates_1960_2014.csv")
print ('original len = ',df.shape[0])
# Assume df is your main DataFrame
# Randomly sample 50% for Group B (Treatment)
group_b = df.sample(n=10, random_state=42)
df.drop(group_b.index).sample(n=10, random_state=1)
print ('len = ',df.shape[0])
```
